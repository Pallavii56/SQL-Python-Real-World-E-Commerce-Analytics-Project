```python
import pandas as pd
```

```python
base_path = "D:/sql-python-analytics/data/raw/archive (1)/"

customers = pd.read_csv(base_path + "olist_customers_dataset.csv")
orders = pd.read_csv(base_path + "olist_orders_dataset.csv")
order_items = pd.read_csv(base_path + "olist_order_items_dataset.csv")
payments = pd.read_csv(base_path + "olist_order_payments_dataset.csv")
products = pd.read_csv(base_path + "olist_products_dataset.csv")
categories = pd.read_csv(base_path + "product_category_name_translation.csv")
```

```python
customers_clean = customers[['customer_id', 'customer_unique_id']].copy()
customers_clean.columns = ['customer_id', 'customer_name']
customers_clean['email'] = customers_clean['customer_name'] + "@email.com"
customers_clean['signup_date'] = "2018-01-01"

customers_clean.to_csv(
    "D:/sql-python-analytics/data/raw/customers.csv",
    index=False
)
```

```python
products = products.merge(
    categories,
    on='product_category_name',
    how='left'
)

products_clean = products[['product_id', 'product_category_name_english']].copy()
products_clean.columns = ['product_id', 'category']
products_clean['product_name'] = products_clean['product_id']

products_clean = products_clean[['product_id', 'product_name', 'category']]

products_clean.to_csv(
    "D:/sql-python-analytics/data/raw/products.csv",
    index=False
)
```

```python
orders_clean = orders[['order_id', 'customer_id', 'order_purchase_timestamp', 'order_status']].copy()
orders_clean.columns = ['order_id', 'customer_id', 'order_date', 'order_status']
orders_clean['order_date'] = pd.to_datetime(orders_clean['order_date']).dt.date

orders_clean.to_csv(
    "D:/sql-python-analytics/data/raw/orders.csv",
    index=False
)
```

```python
order_items_clean = order_items[['order_id', 'product_id', 'order_item_id', 'price']].copy()
order_items_clean['quantity'] = 1

order_items_clean = order_items_clean[
    ['order_item_id', 'order_id', 'product_id', 'quantity', 'price']
]

order_items_clean.to_csv(
    "D:/sql-python-analytics/data/raw/order_items.csv",
    index=False
)
```

```python
payments_clean = payments[['order_id', 'payment_sequential', 'payment_value', 'payment_type']].copy()
payments_clean.columns = ['order_id', 'payment_id', 'amount', 'payment_method']
payments_clean['payment_date'] = "2018-01-01"

payments_clean = payments_clean[
    ['payment_id', 'order_id', 'payment_date', 'amount', 'payment_method']
]

payments_clean.to_csv(
    "D:/sql-python-analytics/data/raw/payments.csv",
    index=False
)
```
