```python
import pandas as pd
import os

# Correct path to CSV folder
data_path = r"../data/raw"

# List CSV files
csv_files = [f for f in os.listdir(data_path) if f.endswith('.csv')]
print("Found CSVs:", csv_files)

# Load all CSVs into a dictionary
dfs = {file.replace('.csv',''): pd.read_csv(os.path.join(data_path, file)) for file in csv_files}

# Quick check
print("Loaded datasets:", dfs.keys())
```

```python
# -----------------------------
# 📝 Quick overview of all datasets
# -----------------------------
for name, df in dfs.items():
    print(f"\n===== Dataset: {name} =====")
    print("Shape:", df.shape)
    print("Columns:", df.columns.tolist())
    print("\nInfo:")
    print(df.info())
    print("\nFirst 5 rows:")
    display(df.head())
    print("\nMissing values per column:")
    print(df.isna().sum())
    print("\nDuplicates:", df.duplicated().sum())

# -----------------------------
# 📊 Basic EDA Plots (Numeric Columns)
# -----------------------------
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="whitegrid")

for name, df in dfs.items():
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    if numeric_cols:
        print(f"\n=== Numeric Plots for: {name} ===")
        df[numeric_cols].hist(bins=15, figsize=(15, 5), layout=(1, len(numeric_cols)))
        plt.suptitle(f"{name} - Numeric Histograms")
        plt.show()

        if len(numeric_cols) > 1:
            plt.figure(figsize=(8,5))
            sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
            plt.title(f"{name} - Correlation Heatmap")
            plt.show()
```

```python
# Merge orders with customers
orders_customers = dfs['orders'].merge(dfs['customers'], on='customer_id', how='left')

# Merge order items with products
order_items_products = dfs['order_items'].merge(dfs['products'], on='product_id', how='left')

# Merge everything for a full view
full_data = orders_customers.merge(order_items_products, on='order_id', how='left')
full_data = full_data.merge(dfs['payments'], on='order_id', how='left')

# Preview the merged dataset
full_data.head()
```

```python
# -----------------------------
# 💰 Revenue & Sales Analysis
# -----------------------------

# Calculate total price per order item
full_data['total_price'] = full_data['quantity'] * full_data['price']

# Total revenue
total_revenue = full_data['total_price'].sum()
print("💵 Total Revenue:", total_revenue)

# Revenue by product
revenue_by_product = full_data.groupby('product_name')['total_price'].sum().sort_values(ascending=False)
print("\nTop 10 Products by Revenue:")
display(revenue_by_product.head(10))

# Revenue by customer
revenue_by_customer = full_data.groupby('customer_id')['total_price'].sum().sort_values(ascending=False)
print("\nTop 10 Customers by Revenue:")
display(revenue_by_customer.head(10))

# Revenue over time (if 'order_date' exists)
if 'order_date' in full_data.columns:
    full_data['order_date'] = pd.to_datetime(full_data['order_date'])
    revenue_over_time = full_data.groupby(full_data['order_date'].dt.to_period('M'))['total_price'].sum()
    print("\nRevenue Over Time:")
    display(revenue_over_time)

    # Plot revenue over time
    revenue_over_time.plot(kind='line', figsize=(12,5), marker='o')
    plt.title("Monthly Revenue Over Time")
    plt.xlabel("Month")
    plt.ylabel("Revenue")
    plt.show()

# -----------------------------
# 🛒 Top-selling Products Plot
# -----------------------------
plt.figure(figsize=(12,6))
sns.barplot(x=revenue_by_product.head(10).values, y=revenue_by_product.head(10).index, palette="viridis")
plt.title("Top 10 Products by Revenue")
plt.xlabel("Revenue")
plt.ylabel("Product")
plt.show()
```

```python
import pandas as pd

# Make sure order_date is datetime
full_data['order_date'] = pd.to_datetime(full_data['order_date'])

# Snapshot date (for recency)
snapshot_date = full_data['order_date'].max() + pd.Timedelta(days=1)

# RFM calculation
rfm = full_data.groupby('customer_id').agg({
    'order_date': lambda x: (snapshot_date - x.max()).days,  # Recency
    'order_id': 'nunique',                                   # Frequency
    'total_price': 'sum'                                     # Monetary
}).reset_index()

# Rename columns
rfm.rename(columns={'order_date':'Recency', 'order_id':'Frequency', 'total_price':'Monetary'}, inplace=True)

# Preview
rfm.head()
```

```python
# Assign quartiles (1=best)
rfm['R_quartile'] = pd.qcut(rfm['Recency'], 4, labels=[4,3,2,1])
rfm['F_quartile'] = pd.qcut(rfm['Frequency'].rank(method='first'), 4, labels=[1,2,3,4])
rfm['M_quartile'] = pd.qcut(rfm['Monetary'], 4, labels=[1,2,3,4])

# Combine RFM score
rfm['RFM_Score'] = rfm['R_quartile'].astype(str) + rfm['F_quartile'].astype(str) + rfm['M_quartile'].astype(str)

# Preview top 5
rfm.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,4))
sns.histplot(rfm['Recency'], bins=30, kde=True)
plt.title("Recency Distribution")
plt.show()

plt.figure(figsize=(12,4))
sns.histplot(rfm['Frequency'], bins=30, kde=True)
plt.title("Frequency Distribution")
plt.show()

plt.figure(figsize=(12,4))
sns.histplot(rfm['Monetary'], bins=30, kde=True)
plt.title("Monetary Distribution")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# -----------------------------
# 1️⃣ Ensure total_price exists
# -----------------------------
full_data['total_price'] = full_data['quantity'] * full_data['price']

# -----------------------------
# 2️⃣ Top 10 Products by Revenue
# -----------------------------
top_products = full_data.groupby('product_name')['total_price'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_products.values, y=top_products.index, palette="viridis")
plt.title("Top 10 Products by Revenue")
plt.xlabel("Revenue")
plt.ylabel("Product")
plt.tight_layout()
plt.savefig("top_products.png", dpi=300, bbox_inches='tight')
plt.show()

# -----------------------------
# 3️⃣ Revenue Over Time
# -----------------------------
full_data['order_date'] = pd.to_datetime(full_data['order_date'])
revenue_time = full_data.groupby(full_data['order_date'].dt.to_period('M'))['total_price'].sum()
revenue_time.index = revenue_time.index.to_timestamp()

plt.figure(figsize=(12,5))
sns.lineplot(x=revenue_time.index, y=revenue_time.values, marker='o')
plt.title("Revenue Over Time")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("revenue_over_time.png", dpi=300, bbox_inches='tight')
plt.show()

# -----------------------------
# 4️⃣ RFM Calculation & Segmentation
# -----------------------------
snapshot_date = full_data['order_date'].max() + pd.Timedelta(days=1)

rfm = full_data.groupby('customer_id').agg({
    'order_date': lambda x: (snapshot_date - x.max()).days,
    'order_id': 'nunique',
    'total_price': 'sum'
}).reset_index()

rfm.rename(columns={'order_date':'Recency', 'order_id':'Frequency', 'total_price':'Monetary'}, inplace=True)

# Safe RFM scoring (handles duplicates)
rfm['R_score'] = pd.qcut(rfm['Recency'], 5, labels=range(5,0,-1))
rfm['F_rank'] = rfm['Frequency'].rank(method='first')
rfm['M_rank'] = rfm['Monetary'].rank(method='first')

rfm['F_score'] = pd.qcut(rfm['F_rank'], 5, labels=range(1,6))
rfm['M_score'] = pd.qcut(rfm['M_rank'], 5, labels=range(1,6))
rfm.drop(['F_rank','M_rank'], axis=1, inplace=True)

rfm['RFM_Score'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)

# -----------------------------
# 5️⃣ RFM Heatmap (Sample: first 50 customers)
# -----------------------------
plt.figure(figsize=(12,6))
sns.heatmap(rfm[['Recency','Frequency','Monetary']].head(50), annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("RFM Heatmap (Sample: First 50 Customers)")
plt.tight_layout()
plt.savefig("rfm_heatmap.png", dpi=300, bbox_inches='tight')
plt.show()

# -----------------------------
# 6️⃣ RFM Segments Count
# -----------------------------
plt.figure(figsize=(12,6))
sns.countplot(y='RFM_Score', data=rfm, 
              order=rfm['RFM_Score'].value_counts().index, 
              palette="YlGnBu")
plt.title("RFM Segments Count")
plt.xlabel("Number of Customers")
plt.ylabel("RFM Score")
plt.tight_layout()
plt.savefig("rfm_segments.png", dpi=300, bbox_inches='tight')
plt.show()

print("✅ All 4 visuals exported: top_products.png, revenue_over_time.png, rfm_heatmap.png, rfm_segments.png")
```
